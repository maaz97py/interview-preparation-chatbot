```python
from transformers import pipeline
from datasets import load_dataset
import soundfile as sf
import torch
```

```python
torch.cuda.is_available()
```

```python
# Use GPU (remove 'device=0' for CPU).
# Used pipline for few lines of code but you can use processor + generate code
synthesiser = pipeline("text-to-speech", "microsoft/speecht5_tts", device=0)

# processor = SpeechT5Processor.from_pretrained("microsoft/speecht5_tts")
# model = SpeechT5ForTextToSpeech.from_pretrained("microsoft/speecht5_tts")
# model = model.to_device("cuda")
# vocoder = SpeechT5HifiGan.from_pretrained("microsoft/speecht5_hifigan")

# speaker embedding
embeddings_dataset = load_dataset("Matthijs/cmu-arctic-xvectors", split="validation")
speaker_embedding = torch.tensor(embeddings_dataset[7306]["xvector"]).unsqueeze(0)
```

```python
t = """I apologize for the confusion. As of my last update in January 2022, OpenAI's models, including Whisper, were not available through the Hugging Face transformers library.
        If you're specifically looking for OpenAI's Whisper model, you should check OpenAI's official documentation or resources for the correct way to use it.
      OpenAI might provide its models through its own API or other means.For the most accurate and up-to-date information"""
```

```python
%%time
speech = synthesiser(t, forward_params={"speaker_embeddings": speaker_embedding})
# sf.write("speech.wav", speech["audio"], samplerate=speech["sampling_rate"])
```
